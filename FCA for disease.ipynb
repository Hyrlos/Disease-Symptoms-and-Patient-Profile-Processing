```python
import pandas as pd
```

## Dataset

```python
df_complet = pd.read_csv("dataset/Disease_symptom_and_patient_profile_dataset.csv")
df_complet.sort_values(["Disease"])
```

```python
df_complet = df_complet.drop_duplicates()
df_complet = df_complet.drop("Gender", axis=1)
df_complet.sort_values(["Disease"])
```

## Data processing

### Age threshold

```python
youngmedium = 25
mediumold = 65

df_complet["Young"] = "No"
df_complet["Average Age"] = "No"
df_complet["Old"] = "No"

df_complet.loc[df_complet['Age'] < youngmedium, 'Young'] = 'Yes'
df_complet.loc[(df_complet['Age'] >= youngmedium) & (df_complet['Age'] <= mediumold), 'Average Age'] = 'Yes'
df_complet.loc[df_complet['Age'] > mediumold, 'Old'] = 'Yes'
df_complet.drop("Age", axis=1, inplace = True)
df_complet
```

### Blood Pressure

```python
df_complet["Low Blood Pressure"] = "No"
df_complet["High Blood Pressure"] = "No"

df_complet.loc[df_complet['Blood Pressure'] == "Low", 'Low Blood Pressure'] = 'Yes'
df_complet.loc[df_complet['Blood Pressure'] == "High", 'High Blood Pressure'] = 'Yes'

df_complet.drop("Blood Pressure", axis=1, inplace = True)
df_complet.sort_values(["Disease"])
```

### Cholesterol Level

```python
df_complet["Low Cholesterol Level"] = "No"
df_complet["High Cholesterol Level"] = "No"

df_complet.loc[df_complet['Cholesterol Level'] == "Low", 'Low Cholesterol Level'] = 'Yes'
df_complet.loc[df_complet['Cholesterol Level'] == "High", 'High Cholesterol Level'] = 'Yes'

df_complet.drop("Cholesterol Level", axis=1, inplace = True)
df_complet.sort_values(["Disease"])
```

### remove Negative Outcome Variable

```python
df_complet.drop(df_complet[df_complet['Outcome Variable'] == "Negative"].index, inplace = True)
df_complet.drop("Outcome Variable", axis=1, inplace = True)
```

### Remove no homogeneous disease:
Disease present several times with différents symptom

```python
listDuplicate = []
for name,group in df_complet.groupby(by="Disease"):
    if len(group) > 1:
        df_complet = df_complet[~df_complet['Disease'].isin([name])]
```

### Processed Dataframe

```python
df_complet = df_complet.set_index("Disease")
```

```python
df_complet.replace({'Yes': 'x', 'No': ''}, regex=False, inplace=True)
```

```python
df_complet.to_csv("FC_disease.csv")
```
